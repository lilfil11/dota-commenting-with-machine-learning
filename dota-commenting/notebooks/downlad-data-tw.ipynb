```python
from moviepy.editor import VideoFileClip

def split_video(input_video, fragment_duration):
    video = VideoFileClip(input_video)
    total_duration = video.duration
    
    start_time = 0
    fragment_number = 4995
    while start_time < total_duration:
        # Создаём папку с фрагментом
        os.mkdir(f'./{fragment_number}')
        os.chdir(f'./{fragment_number}')
        os.mkdir(f'./frames')

        # Вырезаем фрагмент
        end_time = min(start_time + fragment_duration, total_duration)
        fragment = video.subclip(start_time, end_time)
        
        # Сохраняем в файл
        output_filename = f'video.mp4'
        fragment.write_videofile(output_filename, codec='libx264')
         
        fragment_number += 1
        start_time = end_time

        # Возвращаемся в корневую директорию
        os.chdir(f'./..')

    # Закрываем видео
    video.close()


split_video(VIDEO_FILENAME, 10)
```

```python
import ffmpeg
import cv2
import numpy as np

from faster_whisper import WhisperModel
from PIL import Image

# Класс для обработки видео
class Video: 
    def __init__(self, filename):
        self.video_filename = filename
        self.audio_filename = f'audio.wav'

    def extract_audio(self):
        stream = ffmpeg.input(self.video_filename)
        stream = ffmpeg.output(stream, self.audio_filename)
        ffmpeg.run(stream, overwrite_output=True)

    def transcribe_audio(self, model, frag_num):
        self.extract_audio()
        segments, _ = model.transcribe(self.audio_filename, beam_size=5)
        segments = list(segments)

        with open(f'{frag_num}/labels.txt', 'w') as file:        
            for segment in segments:
                file.write(f'{segment.text}')

        os.remove(f'audio.wav')
        
        # return segments

    def extract_frames(self, num_frames, frag_num):
        frames = []
        cap = cv2.VideoCapture(self.video_filename)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        
        # Выбираем кадры равномерно по всему видео
        frame_indices = np.linspace(0, total_frames-1, num=num_frames, dtype=int)
        frame_num = 1
        
        for idx in frame_indices:
            cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
            ret, frame = cap.read()
            if ret:
                frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                frames.append(frame)
                img = Image.fromarray(frame)
                img.save(f'./{frag_num}/frames/{frame_num}.jpg')
                frame_num += 1

        os.remove(f'{self.video_filename}')
        cap.release()
        
        # return frames
```

```python
# Запускаем
from tqdm import tqdm

model = WhisperModel("small")

frag_num = 4995
for i in tqdm(range(341)):
    video_filename = f'{frag_num}/video.mp4'
    video = Video(video_filename)
    video.transcribe_audio(model, frag_num)
    video.extract_frames(20, frag_num)

    frag_num += 1
```
